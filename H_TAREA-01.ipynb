```python
def crear_archivo(path, contenido):
    try: 
        archivo_abierto= open (path, "a")
        archivo_abierto.write(f"\n {contenido}")
        archivo_abierto.close()
    except error:
        print(error)
        print("=(Hubo error")
path = "./new_data_pok.txt"
contenido =new_data_pok
crear_archivo(path,contenido)
```
